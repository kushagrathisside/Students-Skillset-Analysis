# **Endoing Techniques to follow**

**Binary Encoder** needed in *gender, dropper, firstgrad, volunteer, clubs, software, hardware, ever_pitched*

**Label Encoder** (Ordinal Data) needed in *year, branch, type_sch10, per_sch10,loc_sch10, type_sch12, per_sch12, loc_sch12, no_of_lang, dsa, github, communities, competitions*

**One Hot Encoding/User Defined Function** needed in *branch, domain, add_Skills, achievements*

Reference: https://www.geeksforgeeks.org/feature-encoding-techniques-machine-learning/

# Binary Encoding Function

```python
def bin_Encoder(colname : str):
  print("Unique Values in %s are " % colname)
  print(sorted(df[colname].unique()))
  df[colname] = df[colname].apply(
    lambda x: 1 if x == sorted(df[colname].unique())[0] else (0 if x == sorted(df[colname].unique())[1] else None))
  print("Unique Values in %s are " % colname)
  print(sorted(df[colname].unique()))
```

```python
columns_for_binary=['gender', 'dropper', 'firstgrad', 'volunteer', 'clubs', 'software', 'hardware', 'ever_pitched']
for i in columns_for_binary:
  bin_Encoder(i)
  print('\n')
```
